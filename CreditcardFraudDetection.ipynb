Build a sample dataset containing similar distribution of normal and fraudlend transactions.

Number of Fraudlend transaction is 492

```python
legit_sample = legit.sample(n= 244)
```

 1.above will give 244 legit sample

 2.now we will concatenate both legit and fraud sample

```python
 #concat() is used for concatenation

new_dataset = pd.concat([legit_sample,fraud],axis =0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training and Logistic Regression with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy  score on training data:',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy  score on test data:',test_data_accuracy)
```
